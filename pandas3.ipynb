```python
#conditional selection
import numpy as np
import pandas as pd
np.random.seed(101)
df = pd.DataFrame(np.random.randn(5,4), ["A", "B", "C", "D", "E"], ["W", "X", "Y", "Z"])
```

```python
df
```

```python
df > 0
```

```python
df[df > 0]
```

```python
df["W"] > 0
```

```python
df[df["W"] > 0]
```

```python
#multiple conditions
df[(df["W"] > 0) & (df["Y"] < 0)]
```

```python
df[(df["W"] > 0) | (df["Z"] < 0)]
```

```python
#reset index -> sets index to numbers and existing index as a column (not inplace)
df.reset_index()
```

```python
newIdx = ["MH", "GJ", "UP", "KN", "MP"]
df["states"] = newIdx
df
```
